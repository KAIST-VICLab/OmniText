# OmniText demo

A minimal end-to-end example: run **text removal** on a single OmniText-Bench image and display the result.

**Prerequisites** (see the project README):
1. `pip install -e .` (and PyTorch for your CUDA version)
2. `python scripts/download_weights.py`
3. `python scripts/download_data.py`

Removal is the first stage of the OmniText pipeline; editing / repositioning / rescaling consume its `removal.png` output, while insertion uses the original input image.

```python
# Run text removal on the first image only (--limit 1).
# Adjust --gpu / paths as needed.
!python ../scripts/run_removal.py --config ../configs/removal.yaml \
    --dataset-root ../OmniText-Bench --output-dir ../experiments/removal --limit 1
```

```python
import json
from pathlib import Path
import matplotlib.pyplot as plt
from PIL import Image

from omnitext import data as omnidata

dataset_root = Path('../OmniText-Bench')
removal_dir = Path('../experiments/removal')

labels = omnidata.load_labels(dataset_root)
key = next(iter(labels))

original = Image.open(omnidata.input_image_path(dataset_root, key)).convert('RGB')
removed = Image.open(omnidata.removal_output_path(removal_dir, key)).convert('RGB')

fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(original); axes[0].set_title(f'Input ({key})'); axes[0].axis('off')
axes[1].imshow(removed); axes[1].set_title('Text removed'); axes[1].axis('off')
plt.tight_layout(); plt.show()
```
